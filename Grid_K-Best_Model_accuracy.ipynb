```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from xgboost import XGBClassifier
import xgboost as xgb
from sklearn.svm import SVC

import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.feature_selection import SelectKBest, f_classif
```

```python
df = pd.read_csv('features_30_sec.csv')
```

```python
X = df.drop(columns=["label"])
Y = df["label"]
```

```python
# Load the dataset

music_df = pd.read_csv("features_30_sec.csv")

# Remove the file name feature
music_df = music_df.drop(columns=["filename"])

#select the feature columns (Excluding the label column)
X = music_df.iloc[:, :-1]

# Separate the features and target variable
# X = music_df.drop(columns=["label"])
# y = music_df["label"]

# Scale the features using StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

# Apply K -best feature selection by taking k value randomly

```python
# Before applying K-best feature selection, you need to ensure that all
# the features in your dataset are numeric. If you have categorical or text features, 
# you will need to convert them to numeric 
# features using techniques such as one-hot encoding or label encoding.
```

```python
k = 10

selector = SelectKBest(score_func = f_classif, k= k)

X_kbest = selector.fit_transform(X_scaled, music_df["label"])
```

```python
#Train test split
X_train,X_test, Y_train,Y_test = train_test_split(X_kbest, music_df['label'], test_size = 0.2, random_state = 42)
```

```python
# Now use any classifier - Decision Treee, Random Forest
# We can use SVM, Logistic Regression, KNN

rfc = RandomForestClassifier(n_estimators = 50000, max_depth = 10, random_state = 100)
rfc.fit(X_train, Y_train)

# Now we need to predict for tes variable on the trained model
pred = rfc.predict(X_test)

# Accuracy of RandomForestClassifier for K-Best features
accuracy_rfc = accuracy_score(Y_test, pred)
print(accuracy_rfc)
```

```python
#Now we use KNN classifier for K-Best selected feautures
# And will check the accuracy score of KNN

knn = KNeighborsClassifier(n_neighbors = 5)
knn.fit(X_train, Y_train)

# Make predictions on testing set
pred = knn.predict(X_test)
accuracy_knn = accuracy_score(Y_test, pred)
print(accuracy_knn)
```

```python
# Using xgb Classifier

from sklearn.preprocessing import LabelEncoder

# Select K best features

k_best_xgb = SelectKBest(score_func = f_classif, k= 20)
X_kbest = k_best_xgb.fit_transform(X, Y)


X_train, X_test, Y_train, Y_test = train_test_split(X_kbest, Y, test_size = 0.2, random_state = 42)


xgb = XGBClassifier(n_estimators = 1000, learning_rate = 1, max_depth=3)
xgb.fit(X_train, Y_train)

#Prediction on test set
pred = xgb.predict(X_test)
accuracy_Xgb = accuracy_score(Y_test, pred)
print(accuracy_Xgb)
```

```python
# Using xgb Classifier

from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
Y_train_encoded = le.fit_transform(Y_train)
Y_test_encoded = le.fit_transform(Y_test)

# Train XGB Classifier
xgb = XGBClassifier(n_estimators = 50000, learning_rate = 1, max_depth=10)
xgb.fit(X_train, Y_train_encoded)

pred_xgb = xgb.predict(X_test)
accuracy_xgb = accuracy_score(Y_test_encoded, pred_xgb)
print(accuracy_xgb)
```

```python
# Model Training
clf = SVC(kernel = 'rbf', C =5, gamma = 10)
clf.fit(X_train, Y_train)

# Predict
accuracy_svm = clf.score(X_test, Y_test)
print(accuracy_svm)
```

# Now we will be using Grid Search selection for this huge dataset.

```python
# Grid Search selection for SVM classifier
```

```python
# Feature Scaling
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Hyperparameter Tuning
param_grid = {'C': [0.1, 1, 10], 'gamma': [0.1, 1, 10], 'kernel': ['linear', 'rbf', 'poly']}
grid = GridSearchCV(SVC(), param_grid, refit=True, verbose=3)
grid.fit(X_train, Y_train)

print("Best parameters: ", grid.best_params_)
print("Best estimator: ", grid.best_estimator_)

# Training SVM on the training set
svm = SVC(**grid.best_params_)
svm.fit(X_train, Y_train)

# Testing the model on the testing set
accuracy = svm.score(X_test, Y_test)
print("Accuracy: {:.2f}%".format(accuracy * 100))
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# convert the grid search results to a pandas DataFrame
results = pd.DataFrame(grid.cv_results_)

# extract the estimator and accuracy values
estimators = results['param_C']
gammas = results['param_gamma']
accuracies = results['mean_test_score']

# convert the estimator values to numeric type
estimators = pd.to_numeric(estimators, errors='coerce')

# create a pivot table with estimator and gamma as rows and columns, and accuracy as values
pivot_table = pd.pivot_table(data=results, values='mean_test_score', index='param_C', columns='param_gamma')

# plot the heatmap
sns.heatmap(pivot_table, annot=True, cmap='coolwarm', fmt='.3f')

# set plot title and axis labels
plt.title('Estimator and Gamma vs. Accuracy')
plt.xlabel('Gamma')
plt.ylabel('Estimator')

plt.show()
```

```python
# Grid Search for KNN classifier

# Defining parameter grid for grid search
param_grid = {'n_neighbors': [3, 5, 7, 9, 11]}

# Performing grid search with cross-validation
grid = GridSearchCV(KNeighborsClassifier(), param_grid, cv=5)
grid.fit(X_train, Y_train)

# Printing the best parameter and estimator values
print("Best parameters: ", grid.best_params_)
print("Best estimator: ", grid.best_estimator_)

# Training KNN on the training set
knn = KNeighborsClassifier(**grid.best_params_)
knn.fit(X_train, Y_train)

# Testing the model on the testing set
accuracy = knn.score(X_test, Y_test)
print("Accuracy: {:.2f}%".format(accuracy * 100))
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

rfc = RandomForestClassifier(random_state = 42)
rfc.fit(X_train, Y_train)

param_grid = {'n_estimators': [100, 200, 300, 500], 'max_depth': [None, 5, 10, 20]}

grid_search = GridSearchCV(estimator = rfc, param_grid = param_grid, cv = 5, n_jobs = -1, verbose = 2)
grid_search.fit(X_train, Y_train)

# BEST PARAMETERS AND BEST ESTIMATORS
print("Best parameters: ", grid_search.best_params_)
print("Best estimators: ", grid_search.best_estimator_)

accuracy_rfc_grid = rfc.score(X_test, Y_test)
print("Accuracy = {:.2f}%".format(accuracy * 100))
```

```python
# grid search selection using XGBoost classifier

# Feature Scaling

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Convert the target variable to numeric labels

Y_test = np.array(Y_test).astype(str)
label_encoder = LabelEncoder()
Y_train = label_encoder.fit_transform(Y_train)

y_test = label_encoder.transform(Y_test)


# Hyperparameter Tuning using Grid Search
param_grid = {
    'n_estimators': [50, 100, 200], 
    'max_depth': [3, 5, 7], 
    'learning_rate': [0.05, 0.1, 0.01, 0.001, 0.15]
}


grid_search = GridSearchCV(
    XGBClassifier(random_state=42),
    param_grid,
    cv=5,
    n_jobs=-1
)

# Printing best parameters and best estimator
print("Best parameters: ", grid_search.best_params_)
print("Best estimator: ", grid_search.best_estimator_)

# Training and testing XGBoost model with best parameters
xgb_model =   XGBClassifier(**grid_search.best_estimator_.get_params())
xgb_model.fit(X_train, Y_train)

accuracy = xgb_model.score(X_test, Y_test)
print("Accuracy: {:.2f}%".format(accuracy * 100))
```
